```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z ]', '', text)   # remove numbers & punctuation
    text = re.sub(r'\s+', ' ', text)         # remove extra spaces
    return text

df["Cleaned_Text"] = df["News Headline"].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    max_features=1000,
    stop_words='english'
)

X = vectorizer.fit_transform(df["Cleaned_Text"])

print("Shape of TF-IDF matrix:", X.shape)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

subset = X[:500].toarray()

linked = linkage(subset, method='ward')

plt.figure(figsize=(10,6))
dendrogram(linked)
plt.title("Dendrogram (First 500 Articles)")
plt.xlabel("Articles")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(
    n_clusters=5,
    linkage='ward'
)

y_hc = model.fit_predict(X.toarray())

df["Cluster"] = y_hc
```
